### Separating source and target variables

### Checking for imbalanced dataset

Ratio of '0' and '1' is 7.5 so this is an imbalanced dataset.

```python
y.value_counts()
```

### Balancing the imbalanced dataset

Balancing the imbalanced dataset using SMOTE (Synthetic Minority Over sampling Technique). Over sampling minority and under sampling majority values.

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

```python
over = SMOTE(sampling_strategy=0.23)
under = RandomUnderSampler(sampling_strategy=0.5)
```

```python
steps = [('o',over),('u',under)]
pipeline = Pipeline(steps=steps)
```

```python
X,y = pipeline.fit_resample(X,y)
```

```python
X.shape
```

```python
y.value_counts()
```

### Creating model

```python
import lightgbm as lgt
```

```python
model = lgt.LGBMClassifier()
```

```python
model.fit(X,y)
```

### Predicting the output values for test dataset

```python
final = model.predict(t_df)
```

```python
type(final)
```

```python
f= pd.DataFrame(final,columns=['labels'])
```
